```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Cargar la imagen en escala de grises
imagen = cv2.imread('image1.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
# Crear el detector/describidor KAZE
# extended=True para descriptor de 128 bytes, upright=False para rotación invariante
kaze = cv2.KAZE_create(extended=True, upright=False,
                       threshold=0.001, nOctaves=4, nOctaveLayers=4,
                       diffusivity=cv2.KAZE_DIFF_PM_G2)
```

```python
# Detectar puntos clave y calcular descriptores
puntos_clave, descriptores = kaze.detectAndCompute(imagen, None)
```

```python
# Dibujar puntos clave sobre la imagen
imagen_claves = cv2.drawKeypoints(imagen, puntos_clave, None,
                                  flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
```

```python
print(f"Número de puntos clave detectados: {len(puntos_clave)}")
print(f"Dimensión de cada descriptor: {descriptores.shape[1]} bytes")
```

```python
# Mostrar resultados
plt.figure(figsize=(8, 6))
plt.imshow(imagen_claves, cmap='gray')
plt.title('Puntos clave detectados con KAZE')
plt.axis('off')
plt.show()
```
